# Modelling

```python
# Analysis
import pandas as pd
import numpy as np

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

## Linear Models

```python
houses = {'area' : [40, 50, 60, 70, 80], 
          'price': [120000, 150000, 200000, 
                    250000, 300000]}
df = pd.DataFrame(houses)
plt.xlabel('Area (metres sq.)')
plt.ylabel('Price (€)')
sns.scatterplot(data=df, x='area', y='price')
plt.show()
```

```python
sns.lmplot(x='area',y='price',data=df)
plt.xlabel('Area (metres sq.)')
plt.ylabel('Price (€)')
plt.show()
```

## Correlation

### Joint Plot

We can get a quick overview of how each feature varies relative to each other. 

```python
penguins = sns.load_dataset('penguins')
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='flipper_length_mm',kind='reg')
```

### Pair Plot

A pairplot will plot each numerical feature against one another as a matrix

```python
sns.pairplot(penguins, hue='island', kind='reg');
```

Using the `corr()` method, we can get measurements of how each feature are correlated to one another. This is then plotted on a heatmap.

```python
sns.heatmap(penguins.corr(), annot=True, cmap='Blues');
```

```python
tips = sns.load_dataset('tips')
sns.heatmap(tips.corr(), annot=True, cmap='Blues');
```
